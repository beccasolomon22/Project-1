```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#  REVENUE CSV FILES
netflixrev = "/Users/sabrinafernandez/Downloads/NetflixRevenue.csv"
netflixsub = "/Users/sabrinafernandez/Downloads/NetflixSubscribers.csv"
tiktokrev = "/Users/sabrinafernandez/Downloads/TikTokRevenue.csv"
tiktoksub = "/Users/sabrinafernandez/Downloads/TikTokUsers.csv"
ytrev = "/Users/sabrinafernandez/Downloads/YoutubeRevenue.csv"
ytsub = "/Users/sabrinafernandez/Downloads/YoutubePremiumSubscribers.csv"
hulurev = "/Users/sabrinafernandez/Downloads/HuluRevenue.csv"
hulusubs = "/Users/sabrinafernandez/Downloads/HuluNumSubscribers.csv"
Twitch_rev = "/Users/sabrinafernandez/Downloads/TwitchRevenue.csv"
Twitch_Concurrent_View = "/Users/sabrinafernandez/Downloads/TwitchConcurrentViewers.csv"

netflix_rev = pd.read_csv(netflixrev)
netflix_sub = pd.read_csv(netflixsub)
tiktok_rev = pd.read_csv (tiktokrev)
tiktok_sub = pd.read_csv(tiktoksub)
yt_rev = pd.read_csv (ytrev)
yt_sub = pd.read_csv(ytsub)
hulu_rev = pd.read_csv(hulurev)
hulu_sub = pd.read_csv(hulusubs)
Twitch_rev= pd.read_csv(Twitch_rev)
Twitch_Concurrent_View = pd.read_csv(Twitch_Concurrent_View)

netflixmer = pd.merge(netflix_rev, netflix_sub, how="left")
tiktokmer = pd.merge(tiktok_rev, tiktok_sub, how="left")
ytmer = pd.merge(yt_rev, yt_sub, how="left")
hulu_mer = pd.merge(hulu_rev, hulu_sub, how="left")
Twitchmer = pd.merge(Twitch_rev, Twitch_Concurrent_View, how="left")

# Display the data table for preview
netflixmer.head()
```

```python
CompanyByAge = "/Users/sabrinafernandez/Downloads/CompanyByAge.csv.csv"
Company_By_Age = pd.read_csv(CompanyByAge)
Company_By_Age.head()
```

```python
Name = Company_By_Age["Company Name"].tolist()
YearFounded = Company_By_Age["Year Founded"].tolist()
Age = Company_By_Age["Age"].tolist()


fig, ax = plt.subplots(figsize=(12,6))
Company_By_Age.plot(x="Company Name", y="Age", kind="bar", title="Company By Age", ylabel = 'Age of Company in Years', color = ["cyan", "Indigo", "black", "Green", "red"], ec="black", alpha=0.9,
               width=0.4, align="center", ax=ax)
ax.set_facecolor('xkcd:white')
ax.legend().remove()
```
