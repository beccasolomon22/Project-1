```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# REVENUE CSV FILES
yearlycsv = "/Users/sabrinafernandez/Downloads/screentime2022to2023.csv"

# Read the data and the study results
yearlydifference = pd.read_csv(yearlycsv)

# Display the data table for preview
yearlydif = yearlydifference.dropna(how='any')
yearlydif.head()

yearlyclean = yearlydif.sort_values("Difference (Minutes)", ascending=True)
yearlyclean

# Set the figure size before creating the bar chart
plt.figure(figsize=(12, 6))
plt.title('Difference in Yearly Screentime (2022 to 2023)')
plt.xlabel('Country')
plt.ylabel('Difference in Time (Minutes)')

x_axis = yearlyclean["Country"]
y_axis = yearlyclean["Difference (Minutes)"]

plt.bar(x_axis, y_axis, color=['red', 'orange', 'yellow', 'green', 'blue',
                               'indigo', 'violet', 'pink', 'brown', 'black',
                               'magenta', 'cyan', 'gray', 'gold'], align="center")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
