```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#  REVENUE CSV FILES
yearlycsv = "/Users/sabrinafernandez/Downloads/screentime2022to2023.csv"

# Read the mouse data and the study results
yearlydifference = pd.read_csv(yearlycsv)

# Display the data table for preview
yearlydif = yearlydifference.dropna(how='any')
yearlydif.head()
```

```python
yearlyclean = yearlydif.sort_values("Difference (Minutes)", ascending=True)
yearlyclean
```

```python
plt.title('Diffence in Yearly Screentime (2022 to 2023)')
plt.xlabel('Country')
plt.ylabel('Difference (Minutes)')
x_axis=yearlyclean["Country"]
y_axis=yearlyclean["Difference (Minutes)"]
plt.bar(x_axis,y_axis, color=['black', 'red', 'green', 'blue', 'cyan'], align="center")
plt.xticks(rotation=90)
# set_figheight(40)
# set_figwidth(40)
plt.figure(figsize=(2, 4))
plt.tight_layout()
plt.show()
```
